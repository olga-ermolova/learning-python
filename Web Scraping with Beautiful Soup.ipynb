```python
import requests
# dir(requests)
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
import webbrowser as wb
import datetime
import os
```

```python
url = 'https://docs.python.org/3/library/random.html'

# open url to check the page
# wb.open(url)
print("The url was opened in another tab at " + str(datetime.datetime.now()))


# get the page html and ensure it went OK
pageHtml = requests.get(url)
if pageHtml.status_code == 200: # or use pageHtml.raise_for_status()
    print("The request for HTML code succeeded")
else:
    print("The request for HTML code failed, error code: " + str(pageHtml.status_code))

htmlLength = len(pageHtml.text)
print("The HTML length is " + str(htmlLength) + " characters")


# now work with the html
# get the function names first
soup = bs(pageHtml.text, 'html.parser')
# print(soup)
names = soup.body.findAll('dt') # get all elements with 'dt' tag = description
# print(names)
function_names = re.findall('id="random.\w+', str(names))
# print(function_names)

fn = list()
for i in function_names:
    fn.append(i[4:])
print("\nHere are the functions in a 'random' module:\n", fn, '\n')    


# now get the function descriptions
descrRaw = soup.body.findAll('dd')
# print(descrRaw)
descr = list()
for i in descrRaw:
    i = i.text.replace('\n', ' ')
    descr.append(i)
# print("\nHere go the descriptions:\n", descr, '\n')

print('qty of functions: ', str(len(fn)))
print('qty of descr: ', str(len(descr)), '\n')


# option 1: combine into tuples
zip1 = zip(fn, descr)
# print(tuple(zip1))

# better option: combine into a dictionary
data = pd.DataFrame({'function name': fn, 'function usage': descr})
data 
```
